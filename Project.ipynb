## 04. Data Imputation

Now, let us check the null values across different columns in the dataset

```python
# To check the sum of null values in different columns
data.isnull().sum()
```

Now, we can view the 11 rows which have null entries of Postal Code. Since the count of null values is less, we can try to impute them using appropriate techniques.

```python
# View rows with null Postal Code
data[pd.isnull(data["Postal Code"])]
```

We can clearly see that all rows with Postal Code as null values have city as Burlington and State as Vermont in common. So, all the missing Postal Codes are all same. On further research, we found that the Postal Code of Burlington in Vermont is "05401". So, we can replace the null values with it. 

```python
# Replacing missing values with "05401"
data.loc[(data['City'] == "Burlington") & (data['State'] == "Vermont"),'Postal Code'] = "05401"
```

```python
# Checking the sum of null values again
data.isnull().sum()
```

Here, we can find that none of the columns have null values now. So, now the data is clean.

```python
# Viewing the replaced values
data.loc[(data['City'] == "Burlington") & (data['State'] == "Vermont"),]
```

## 06. Influence of Categorical Variables

Let us try to create a pivot table for each of the categorical variables and calculate the respective profitability ratio to compare the variable's influence on the profitability. 

We did not calculate the profitability ratio on the whole data, we tried to aggregate the data to the required Level of Detail and then calculate the aggregation measure. 

### (i) State

```python
# Aggregating data by state
group_by_state = data.groupby(by='State').agg(**{'sales_sum':('Sales','sum'), 'profit_sum':('Profit','sum')})
group_by_state['profit_ratio'] = group_by_state['profit_sum'].div(group_by_state['sales_sum'])
```

```python
# Printing grouped data
group_by_state
```

```python
# Arranging the data in descending order
group_by_state.sort_values("profit_ratio", ascending=False)
```

```python
# Plotting the data between states and their respective profitability ratio
sns.set(rc={'figure.figsize':(18,9)})
sns.barplot(x=group_by_state.index, y="profit_ratio", data=group_by_state)
plt.xticks(rotation=90)
plt.show()
```

You can see here that the profitability ratio is very high in states like Connecticut, Delaware etc. whereas it is negative in states like Ohio, Coloroda etc. 

This explains that state has an influence on the profitability ratio. 

```python
group_by_state.sort_values("profit_ratio", ascending=False)
```

```python
# List of states with profit ratio greater than 0.2
group_by_state[group_by_state["profit_ratio"]>0.2].sort_values("profit_ratio", ascending=False).index
```

### (ix) Customer ID

We will try to group the data using ccustomer ID and calculate their specific measures.

```python
# Unique number of Customers
data["Customer ID"].nunique()
```

Totally we have 793 customers

We will create a new dataframe with Customer ID as Identifier and various customer specific features such as First Order date, last order date, date differene, last year order, first year order, number of orders, number of items, total sales, total profita nd profit ratio

```python
users = pd.unique(data['Customer ID'])
ed, rd, dd, yd, ym, no, ni, avsales, avprof, profrat = [], [], [], [], [], [], [], [], [], []

for user in users:
    temp = data.loc[data['Customer ID'] == user]
    avsales.append(temp.Sales.sum())
    avprof.append(temp.Profit.sum())
    profrat.append(temp.Profit.mean()/temp.Sales.mean())
    ed.append(temp["Order Date"].min())
    rd.append(temp["Order Date"].max())
    yd.append(temp["Order Year"].max())
    ym.append(temp["Order Year"].min())
    ni.append(len(temp))
    no.append(len(pd.unique(temp['Order ID'])))
    
for i in range(len(ed)):
    tmp = pd.to_datetime(rd[i])-pd.to_datetime(ed[i])
    dd.append(tmp)
    
customer_data = pd.DataFrame(
    {'Customer ID':users, 
      'First Order':ed, 
      'Last Order':rd, 
      'Date Difference':dd,
      'Year Last Order':yd,
      'Year First Order':ym,
      'Number Orders':no,
      'Number Items':ni,
      'Total Sales': avsales,
      'Total Profit': avprof,
      'Profitability Ratio': profrat})
customer_data.head()
```

```python
# Build Correlation Matrix
corrMatrix1 = customer_data.corr()
sns.heatmap(corrMatrix1, annot=True)
plt.show()
```

Looking at the correlation matrix, we find that Total Profit is correlated with profitability ratio. 

We can also find that Total Profit is positively correlated with Total Sales. In addition to that, Total Sales is highly correlated with Number of Items purchased and Number of Orders. 

We also find that First Order year is negatively correlated with Number of Orders and Number of Items. This is obvious as more recent the customers are the more less their number of orders and number of items purchased. 

```python
# Aggregating data by Customer ID, Plotting the data between Customer ID and their respective profitability ratio
group_by_sub_category = data.groupby(by='Customer ID').agg(**{'sales_sum':('Sales','sum'), 'profit_sum':('Profit','sum')})
group_by_sub_category['profit_ratio'] = group_by_sub_category['profit_sum'].div(group_by_sub_category['sales_sum'])
sns.set(rc={'figure.figsize':(18,9)})
sns.barplot(x=group_by_sub_category.index, y="profit_ratio", data=group_by_sub_category)
plt.xticks(rotation=90)
plt.title('Profit ratio by Customer ID')
plt.show()
```
